# Theory

(i) How does an insurance company decide how much to charge for car insurance?

(ii) How do casinos set up payout structures to make sure that they make a profit?

(iii) How does a poker player decide whether or not to fold?

(iv) How do we efficiently measure the appeal of a political candidate?

The answers to these questions are informed by the Law of Large Numbers.

--> The law of large numbers states that as the number of trials or observations increases, the actual or observed probability approaches the
theoretical or expected probability.

--> Example: Flipping a fair coin.

    P(Head) = 0.5 ; P(Tail) = 0.5
    
    These are nothing but the theoretical probabilities.

    But this does not imply that if there are 10 coin flips,
    we will get 0.5*10 = 5 Heads.

    However, we can be confident that as we continue flipping a coin       indefinitely the cumulative proportion of heads flipped
    should get closer and closer to 50%.

   






 One common misconception is that if the first four flips were heads, you’re more likely to get tails on the next flip. This is not the case because each flip of a coin is an independent event, its outcome is unaffected by all previous events.

So if you start out with four heads in a row, it’s not that you are more likely to get tails, it’s just that in the grand scheme of things, four head flips will get averaged with a huge number of flips that are expected to yield an even number of heads and tails, causing the proportion to approach 50% as the number of trials increases.

The Law of Large numbers explains that the more people from a population that you sample, the closer your sample average will be to the population average. This is assuming that the samples are free from any kind of bias.  




# Code

```python
#importing necessary libraries
import random
import matplotlib.pyplot as plt
import numpy as np

#Setting the seed,  used to initialize the random number generator.
random.seed(1991)

#Defining a function for rolling a dice n times
def Roll_dice(n):
    #List to store the result of dice roll
  result = []
  for i in range(1,n+1):
      result.append(random.choice([1,2,3,4,5,6]))
  return result


#Now, I want to plot how the average of dice rolls converges to the expected value
#Expected value = (X1P1 +X2P2 +..... +XnPn)
#                = (1*1/6) + (2*1/6) + (3*1/6) + (4*1/6) + (5*1/6) + (6*1/6)
#                = 1/6 + 2/6 + 3/6 + 4/6 + 5/6 + 6/6
#               = 21/6
#               = 3.5

#Now, lets roll the dice for a certain no. of times

def results(n):

  result=   Roll_dice(n)
  averages = []

  _cumsum = np.cumsum(result)
  for index in range(len(_cumsum)):
    averages.append(_cumsum[index]/(index+1))

  return averages
```

```python
avgs = results(10)

#Plotting the averages
plt.plot(avgs)

#Adding information to the graph
plt.xlabel('Trials') #naming the x-axis
plt.ylabel('Average of all outcomes') #naming the y-axis
plt.title('Average value of Dice rolled ') #Graph title
```

```python
avgs = results(100)


plt.plot(avgs)

#Adding information to the graph
plt.xlabel('Trials') #naming the x-axis
plt.ylabel('Average of all outcomes') #naming the y-axis
plt.title('Average value of Dice rolled ') #Graph title
```
